```python
import random

# Define the number of IDs to generate
num_ids = 100

# Define the length of the ID and password
id_length = 10
password_length = 6

# Define the starting number for the ID
id_start = 30

# Generate the IDs and passwords
ids = []
passwords = []

for i in range(num_ids):
    # Generate a random ID number
    id_number = str(id_start) + str(random.randint(10**(id_length-2), 10**(id_length-1)-1))
    ids.append(id_number)
    
    # Generate a random password
    password = str(random.randint(10**(password_length-1), 10**password_length-1))
    passwords.append(password)

# Print the IDs and passwords
for i in range(num_ids):
    print("ID: " + ids[i] + " Password: " + passwords[i])
```
