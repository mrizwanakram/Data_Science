### Lecture 61: Abstract classes and method
PythonServer Side ProgrammingProgramming. A class is called an Abstract class if it contains one or more abstract methods. An abstract method is a method that is declared, but contains no implementation. Abstract classes may not be instantiated, and its abstract methods must be implemented by its subclasses.

```python
#ABC = Abstract Base Clases
```

```python
from abc import ABC, abstractmethod

class computer (ABC):
    @abstractmethod
    def process (self):
        pass
    
class laptop(computer):
    def process(self):
        print("its ranning")

class programmer:
    
    def work(self):
        print("solving bugs")

comp1 =  laptop()

prog1 = programmer()
prog1.work()

comp1.process()
```

### Lecture 63: Genraters
Python provides a generator to create your own iterator function. A generator is a special type of function which does not return a single value, instead, it returns an iterator object with a sequence of values. In a generator function, a yield statement is used rather than a return statement.

```python
def topten():
    
    yield 5
    
values = topten()

print(values)
    
    
```

```python
def topten():
    
    return 5
    
values = topten()

print(values)
    
```

```python
def topten():
    
    yield 1
    yield 2
    yield 3
    yield 4
    
values = topten()

print(values.__next__())
print(values.__next__())

for i in values:
    
    print(i)
    
```

```python
def topten():
    
    n = 1  
    while n <= 10:
        sq = n * n
        yield sq
        n += 1

values = topten()


for i in values:
    
    print(i)
```

### Lecture 65: Multithreading
Multithreading is defined as the ability of a processor to execute multiple threads concurrently. In a simple, single-core CPU, it is achieved using frequent switching between threads.

```python
class Hello:
    def run(self):
        for i in range(5):
            print("Hello")
            
            
class Hi:
    def run(self):
        for i in range(5):
            print("Hi")
            
t1 = Hello()
t2 = Hi()

t1.run()
t2.run()
```

```python

from threading import *

class Hello(Thread):
    def run(self):
        for i in range(5):
            print("Hello")
            
            
            
class Hi(Thread):
    def run(self):
        for i in range(5):
            print("Hi")
            
t1 = Hello()
t2 = Hi()

t1.start()
t2.start()
```

### Lecture 68: Linear search
One of the most straightforward and elementary searches is the sequential search, also known as a linear search. As a real world example, pickup the nearest phonebook and open it to the first page of names. We're looking to find the first "Smith". Look at the first name.

```python
def search(list,n):
    i = 0
    
    while i < len(list):
        if list [i] == n:
            
            return True
        i = i + 1
    return False

list = [2,3,4,5,6,7]
n = 4
if search (list, n):
    print("Found")
else:
    print("Not Found")
```
